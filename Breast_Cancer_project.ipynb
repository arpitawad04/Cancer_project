# Create Model:

We are using here K nearest neighbor (KNN) which is a machine learning algorithm.K-Nearest Neighbour is one of the simplest Machine Learning algorithms based on Supervised Learning technique.
K-NN algorithm assumes the similarity between the new case/data and available cases and put the new case into the category that is most similar to the available categories.
K-NN algorithm can be used for Regression as well as for Classification but mostly it is used for the Classification problems.
Here we have used KNN classifier because of our problem type.

# Split the data in Training and validation:


Let's split the data randomly into training and test set.

We will fit/train a classifier on the training set and make predictions on the test set. Then we will compare the predictions with the known labels.

Scikit-learn provides facility to split data into train and test set using train_test_split method

```python
#importing train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=1)
```

Initially we are splitting our data as 70% training and 30% testing.Using train_test_split function and hold out cross vaidation splitting of data is carried out.Random state kept at 42(it is basically used to obtain same result if we run it again ,in case we dont use this everytime results would be something different)

# Let's create a classifier using k-Nearest Neighbors algorithm.

First let us first observe the accuracies for different values of k.

```python
#import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier

#Neighbors chosen 
neighbors=range(1,23)

#Setup arrays to store training and test accuracies
train_accuracy=np.empty(len(neighbors))
test_accuracy=np.empty(len(neighbors))

```

```python
#For loop is used to train and test model on different value of K(neighbors):
for i,k in enumerate(neighbors):
    
    #Setup a knn classifier with k neighbors
    knn=KNeighborsClassifier(n_neighbors=k)
    
    #Fit the model
    knn.fit(X_train,y_train)
    
    #Predict on test data
    y_pred=knn.predict(X_test)
    
    #Compute accuracy on the training set
    train_accuracy[i]=knn.score(X_train,y_train)
    
    #Compute accuracy on the test set
    test_accuracy[i]=knn.score(X_test,y_test)
```

# Generate Plot:

```python
#Generate plot
plt.title('k-NN Varying number of neighbors')
plt.plot(neighbors,test_accuracy, label='Testing Accuracy')
plt.plot(neighbors,train_accuracy, label='Training accuracy')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```

As we can see that by using 70 to 30 % ratio for splitting data using hold out cross validation ,our results are not as expected.Training accuracy is dropping as we keep on increasing neighbors K.

# Cross Validation:

Now before getting into the details of Hyperparamter tuning, let us understand the concept of Cross validation.

The trained model's performance is dependent on way the data is split. It might not  be representative of the modelâ€™s ability to generalize.

The solution is cross validation.

Cross-validation is a technique to evaluate predictive models by partitioning the original sample into a training set to train the model, and a test set to evaluate it.

The advantage of this method is that all observations are used for both training and validation, and each observation is used for validation exactly once.

# Hyperparameter tuning

Earlier the value of K(=7) was selected on the basis of accuracy vs number of neighbors.This is primitive way of hyper parameter tuning

In Scikit learn there is a simple way of doing Hyperparameter tuning using GridSearchCV(Grid Search Cross validation).
GridSearchCV taking different parameters given to model ata time and find the best possible parameter combination in case of multiple parameters in terms of accuracy.

```python
#import GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
#In case of classifier like knn the parameter to be tuned is n_neighbors
param_grid = {'n_neighbors':np.arange(1,23)}
```

```python
knn = KNeighborsClassifier()
knn_cv= GridSearchCV(knn,param_grid,cv=5)
knn_cv.fit(X_new,y)
```

```python
knn_cv.best_score_
```

```python
knn_cv.best_params_
```

By using GridSearchCV , we obtain that optimum value of k=6 having accuracy score 0.968.
